```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from laplacian_blend import laplacian_blender
```

### Test images and mask

```python
sun = cv2.imread("Cat_face.png")
sun = cv2.cvtColor(sun, cv2.COLOR_BGR2RGB)
moon = cv2.imread("Dog_face.png")
moon = cv2.cvtColor(moon, cv2.COLOR_BGR2RGB)

mask = np.concatenate((np.ones((512,256,3)),np.zeros((512,256,3))), axis=1)
```

### Test blend()

```python
# Instantiate laplacian blender class
blender = laplacian_blender(sun,moon,mask)
numLevels = 6
blended_image = blender.blend(numLevels)

plt.imshow(blended_image)
```

### Test ComputePyr()

```python
gaussianPyramid, laplacianPyramid = blender.ComputePyr(sun,numLevels)

# Normalize pyramids
gaussianPyramid = blender.normalize(gaussianPyramid)
laplacianPyramid = blender.normalize(laplacianPyramid)

for i, img in enumerate(gaussianPyramid):
    # Display the image
    plt.figure(figsize=(5, 5))
    #plt.imshow(img)
    plt.imshow(img, cmap='gray') #use if image is grayscale
    plt.title(f"Gaussian Level: {i}")
    plt.axis('off')  # Turn off axes
    plt.show()

for i, img in enumerate(laplacianPyramid):
    # Display the image
    plt.figure(figsize=(5, 5))
    #plt.imshow(img)
    plt.imshow(img, cmap='gray') #use if image is grayscale
    plt.title(f"Laplacian Level: {i}")
    plt.axis('off')  # Turn off axes
    plt.show()
```
